#### 1. All the Excel workbooks that you want to combine should be placed in same folder.


#### 2.Open a new Excel workbook.


#### 3.Press ALT + F11 (or go to Developer > Visual Basic). This will open the Visual Basic Editor.


#### 4.In the VB Editor menu go to Insert > Module. This will insert a module for the workbook with a blank code window.


#### 5.write below mentioned code in code window.


#### 6. In the code, you need to give the all excel folder location the following line of code:
          Path = ''


#### 7. Place the cursor anywhere in the code and press the F5 key or click on the green play button in the Toolbar.


#### 8.This will activate the code and all worksheets from all the Excel workbooks in the folder would get consolidated into a single workbook. And Save the Master Workbook
